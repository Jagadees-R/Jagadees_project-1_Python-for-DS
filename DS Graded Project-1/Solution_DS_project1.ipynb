# 4. Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix. 
 -  Note: When the record/entry has "M"(indicates millions) as suffix you need to multiply that value with 1000000
 -  When the record/entry has "K"(indicates thousands) as suffix you need to multiply that value with 1000


```python
## removing '€' symbol from ("vaue","Wage","Release Clause") columns.

fifa_df.Value = [x.strip('€') for x in fifa_df.Value]
fifa_df.Wage = [x.strip('€') for x in fifa_df.Wage]
fifa_df['Release Clause'] = [x.strip('€') for x in (fifa_df['Release Clause'].astype(str))]
```

```python
fifa_df
```

```python
## Converting columns("Value","Wage","Release Clause") to float
## Multply 1000000 fro 'M' and 1000 for 'K'

fifa_df.Value = [float(x.strip('M'))*1000000 if 'M' in x else float(x.strip('K'))*1000 for x in fifa_df.Value]
fifa_df.Wage = [float(x.strip('M'))*1000000 if 'M' in x else float(x.strip('K'))*1000 for x in fifa_df.Wage]
fifa_df['Release Clause'] = [float(x.strip('M'))*1000000 if 'M' in x else float(x.strip('K'))*1000000 for x in (fifa_df['Release Clause'].astype(str))]
```

# 5. Convert the column "Joined" into integer data type with keeping only the year.

# 8. The column "Weight" has the suffix as lbs, remove the suffix and convert to float. 

```python
## converting "Weight" column to 'float' datatype anfter removing lbs suffix

fifa_df.Weight = ['0' if 'nan' in x else float(x.strip("lbs")) for x in fifa_df.Weight.astype(str)]
fifa_df.Weight = fifa_df.Weight.astype(float)

fifa_df.Weight
```

# 9. Check for the percentage of missing values and impute them with appropriate imputation techniques. 

```python
## count the total number of entries with missing values - column wise

fifa_df.isnull().sum()
```

```python
## percentage of missing entries - column wise
for i in range(fifa_df.shape[1]):
    col = fifa_df.columns.values.tolist()[i]
    n_miss = fifa_df[col].isna().sum()
    perc = n_miss / fifa_df.shape[0] * 100
    print(' %d  %s , Total-Missing = %d    (%.1f%%)' % (i, col, n_miss, perc))
```

```python
## including imputation for missing entries

## Imputation for column - float dataype
fifa_df[["International Reputation","Weak Foot","Skill Moves","Jersey Number","Crossing","Finishing"]] = fifa_df[["International Reputation","Weak Foot","Skill Moves","Jersey Number","Crossing","Finishing"]].fillna(0)
fifa_df[["HeadingAccuracy","ShortPassing","Volleys","Dribbling","Curve","FKAccuracy","LongPassing"]] = fifa_df[["HeadingAccuracy","ShortPassing","Volleys","Dribbling","Curve","FKAccuracy","LongPassing"]].fillna(0)
fifa_df[["BallControl","Acceleration","SprintSpeed","Agility","Reactions","Balance","ShotPower"]] = fifa_df[["BallControl","Acceleration","SprintSpeed","Agility","Reactions","Balance","ShotPower"]].fillna(0)
fifa_df[["Jumping","Stamina","Strength","LongShots","Aggression","Interceptions","Positioning"]] = fifa_df[["Jumping","Stamina","Strength","LongShots","Aggression","Interceptions","Positioning"]].fillna(0)
fifa_df[["Vision","Penalties","Composure","Marking","StandingTackle","SlidingTackle","GKDiving"]] = fifa_df[["Vision","Penalties","Composure","Marking","StandingTackle","SlidingTackle","GKDiving"]].fillna(0)
fifa_df[["GKHandling","GKKicking","GKPositioning","GKReflexes","Release Clause"]] = fifa_df[["GKHandling","GKKicking","GKPositioning","GKReflexes","Release Clause"]].fillna(0)

## Imputation for column - object datatype
fifa_df[["Club","Preferred Foot","Work Rate","Body Type","Position","Loaned From"]] = fifa_df[["Club","Preferred Foot","Work Rate","Body Type","Position","Loaned From"]].fillna('Missing')

## Imputation for column "Contract Valid Until" - datetime datatype
fifa_df['Contract Valid Until'] = pd.to_datetime(fifa_df['Contract Valid Until'], errors='coerce').ffill()


fifa_df.info()
```

# 15. Generate a dataframe which should include the "Player name", "Club Name", "Wage", and 'Overall rating'.

 ####  -  i) find the average Overall rating for each club.

```python
temp_df = pd.DataFrame.copy(fifa_df[['Name','Club','Wage','Overall']])
temp_df.head(5)
```

```python
avg_club_rat = temp_df.groupby(['Club'])[['Overall']].mean()
avg_club_rat
```

####    - ii) Display the average overall rating of Top10 Clubs using a plot

```python
top_club = {}
for club in temp_df['Club'].unique():
    overall_rating = temp_df['Overall'][temp_df['Club'] == club].mean()
    top_club[club] = overall_rating
    
best_club = pd.DataFrame.from_dict(top_club, orient = 'index', columns = ['overall'])
best_club['club'] = best_club.index
best_club = best_club.sort_values(by = 'overall', ascending =  False)

plt.figure(1 , figsize = (16 , 8))
sns.barplot(x ='club', y ='overall', data = best_club.head(10), palette = 'ocean')  
plt.xticks(rotation = 32)

plt.xlabel('Club', fontsize = 14)
plt.ylabel('Average Overall Rating of players in club', fontsize = 14)
plt.title('Best Players in the Average Overall rating of top 10 clubs', fontsize = 16)
plt.ylim(0 , 100)

plt.show()
print(best_club.head(10))
```
